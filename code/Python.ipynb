```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

mycolors = plt.cm.inferno(range(50))  # 预设的调色板（个人认为这个比较好看，总共有12个）
matplotlib.style.use('ggplot')           # 设置和ggplot类似的风格
sns.set_context('paper',font_scale=2)    # 设置一下参数，以及字体规模
```

```python
fig = plt.figure(figsize=(8,6))    # 定义画布
ax = fig.add_subplot()             # 在画布中定义子图      

for i in range(12):
    plt.plot(i,i,'*',markersize=200,color=mycolors[i])

plt.ylabel('Y')
plt.xlabel('X')
# plt.title()
plt.legend(['1','2'])
plt.tight_layout()
plt.show()
fig.savefig('.pdf')
```
